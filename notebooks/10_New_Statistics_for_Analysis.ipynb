```python
results = df[targets + targets_hat].dropna().reset_index(drop=True)
results
```

```python
chunk_size = 100
num_chunks = len(results) // chunk_size

for i in range(num_chunks):
    start_idx = i * chunk_size
    end_idx = (i + 1) * chunk_size
    chunk_df = results.iloc[start_idx:end_idx]
    # print(start_idx, end_idx)
    # print(chunk_df)
    mse_list = []
    for col, col_hat in zip(targets, targets_hat):
        #mse = mean_squared_error(chunk_df[col], chunk_df[col_hat])
        mse = ((chunk_df[col] - chunk_df[col_hat]) ** 2).mean()
        mse_list.append(mse)
    
    # Print MSE for each target and its _hat variation
    print(f"{100 + (i+1)*chunk_size} "+"-"*50)
    for col, col_hat, mse in zip(targets, targets_hat, mse_list):
        print(f"{100 + (i+1)*chunk_size} - MSE for {col} and {col_hat}: {mse:.2e}")
```
